```python
import random
import time

def partition(arr, low, high):
    i = (low - 1)
    pivot = arr[high]

    for j in range(low, high):
        if arr[j] <= pivot:
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)

        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)

def mergeSort(arr,l,r):
    if l < r:
        m = (l+(r-1)) // 2

        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)

def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = [0] * (n1)
    R = [0] * (n2)

    for i in range(0, n1):
        L[i] = arr[l + i]

    for j in range(0, n2):
        R[j] = arr[m + 1 + j]

    i = 0
    j = 0
    k = l

    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1

    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
        
        
list_Min = 20000
list_Max = 1000000
list_Up = 20000
```

```python
quickSort_List = []

def quickSort_Run():
    for Random in range(list_Min, list_Max + 1, + list_Up):
        total_List = []
        random_Sample = random.sample(range(1, Random + 1), Random)
        partition(random_Sample, 0, len(random_Sample) - 1)
        # print("After partition = ", random_Sample)

        for _ in range(10):
            random_Sample_Copy = random_Sample.copy()
            start_time = time.time()
            quickSort(random_Sample_Copy, 0, len(random_Sample_Copy) - 1)
            end_time = time.time()

            total_time = end_time - start_time
            total_List.append(total_time)

        # print("After quickSort = ", random_Sample_Copy)
        quickSort_List.append(min(total_List))
        # print("quick_List = ", quickSort_List)



quickSort_Run()
print("quick time = ", quickSort_List)
```

```python
mergeSort_List = []

def mergeSort_Run():
    for Random in range(list_Min, list_Max + 1, + list_Up):
        total_List = []
        random_Sample = random.sample(range(1, Random + 1), Random)
        merge(random_Sample, 0, 2, len(random_Sample) - 1)
        # print("After Merge = ", random_Sample)

        for _ in range(10):
            random_Sample_Copy = random_Sample.copy()
            start_time = time.time()
            mergeSort(random_Sample_Copy, 0, len(random_Sample_Copy) - 1)
            end_time = time.time()

            total_time = end_time - start_time
            total_List.append(total_time)

        # print("time = ",total_time)
        # print("After mergeSort = ", random_Sample_Copy)
        mergeSort_List.append(min(total_List))
        # print("quick_List = ", mergeSort_List)



mergeSort_Run()
print("merge time = ", mergeSort_List)
```

```python
import matplotlib.pyplot as plt

# 
x = []
for i in range(list_Min, list_Max + 1, + list_Up):
    x.append(i)

y1 = []
for i in range(0, len(quickSort_List)):
    y1.append(quickSort_List[i])
    
y2 = []
for i in range(0, len(mergeSort_List)):
    y2.append(mergeSort_List[i])

# 
plt.plot(x, y1, label='Quick Sort', color='r', marker='o')
plt.plot(x, y2, label='Merge Sort', color='g', marker='x')

#  x, y 
plt.xlabel('Num List')
plt.ylabel('Time/sec')
plt.title('Advanced Sorting Graph')

#  legend
plt.legend()

# 
plt.show()
```
